```python
lst = []

for i in range(1,1508):
    i = str(i)
    url = 'https://anapioficeandfire.com/api/characters/' +i
    response = requests.get(url)
    got = response.json()
    lst.append(got)
    
for i in range(1512,1995):
    i = str(i)
    url = 'https://anapioficeandfire.com/api/characters/' +i
    response = requests.get(url)
    got = response.json()
    lst.append(got)
    
for i in range(1996,2138):
    i = str(i)
    url = 'https://anapioficeandfire.com/api/characters/' +i
    response = requests.get(url)
    got = response.json()
    lst.append(got)
    
```

```python
def isnameinlist(name):
    for item in lst:
        if name == item['name'] or name in item['aliases']:
            return True
        
name = input("Enter a name: ")
if isnameinlist(name) is True:
    print("Yeah that dude is in Game of Thrones.")
else:
    print('That person is not in Game of Thrones.')



    
```
